```python
import numpy as np
import pandas as pd
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.callbacks import Callback
from sklearn.model_selection import train_test_split
from keras.layers import Dense,Embedding,SimpleRNN
from keras.preprocessing.text import Tokenizer
```

```python
import pandas as pd
df = pd.read_csv("/content/Train.csv")
x = df["text"].values
y = df["label"].values

x_filtered = []

for review in x:
    review = review.lower()
    for i in review:
        punc = '''  !()-[]{};:'"\,<>./?@#$%^&*_~  '''
        if i in punc:
            review = review.replace(i, " ")
    x_filtered.append(review)
```

```python
df_test = pd.read_csv("/content/Test.csv")
xTest = df_test["text"].values
yTest = df_test["label"].values

x_filteredTest = []

for reviewTest in xTest:
    reviewTest = reviewTest.lower()
    for i in reviewTest:
        puncTest = '''  !()-[]{};:'"\,<>./?@#$%^&*_~  '''
        if i in puncTest:
            reviewTest = reviewTest.replace(i, " ")
    x_filteredTest.append(reviewTest)
```

```python
df_val = pd.read_csv("/content/Valid.csv")
xval = df_val["text"].values
yval = df_val["label"].values

x_filteredval = []

for reviewval in xval:
    reviewval = reviewval.lower()
    for i in reviewval:
        puncval = '''  !()-[]{};:'"\,<>./?@#$%^&*_~  '''
        if i in puncval:
            reviewval= reviewval.replace(i, " ")
    x_filteredval.append(reviewval)
```

```python
print(xTest[4])
```

```python
vocab_size = 124253
embed_size = 64
max_len = 50

tokenizer = Tokenizer(num_words=vocab_size,
                      filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n',
                      lower=True,
                      split=" ",
                      oov_token="<OOV>")
```

```python
tokenizer.fit_on_texts(x)
tokenizer.fit_on_texts(xTest)
tokenizer.fit_on_texts(xval)
```

```python
print(tokenizer.word_index)
```

```python
training_tokens = tokenizer.texts_to_sequences(x)
training_tokensTest = tokenizer.texts_to_sequences(xTest)
training_tokensValidation = tokenizer.texts_to_sequences(xval)
```

```python
X_Trainning = pad_sequences(training_tokens, maxlen=max_len, truncating='post')
x_Test = pad_sequences(training_tokensTest, maxlen=max_len, truncating='post')
x_Validation = pad_sequences(training_tokensValidation, maxlen=max_len, truncating='post')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_Test, Y_train, Y_Test = train_test_split(X_Trainning,y,test_size=0.2, random_state=2)
```

```python
embed_size = 64
model = Sequential()
model.add(Embedding(vocab_size, embed_size, input_length=max_len))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
X_train.shape, Y_train.shape, x_Validation.shape, Y_val.shape
```

```python

from keras.optimizers import Adam
lr = 0.001
optimizer = Adam(learning_rate=lr)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['acc'])
```

```python
#history = model.fit(X_train, Y_train, epochs=10, validation_data=(X_val, Y_val))
model.fit(X_train,Y_train, epochs=10,  validation_data=(x_Validation, yval))
```

```python
from keras.models import load_model

# Sauvegarder le modèle
model.save('SentimentAnalyser2.h5')
```

```python

predicting_tokens = tokenizer.texts_to_sequences(["An outstanding piece of cinema. The performances were brilliant, and the storyline was compelling."])
predicting_padded = pad_sequences(predicting_tokens, maxlen=max_len, truncating='post')
prediction = model.predict(predicting_padded)
print(prediction)
rounded_prediction = int(round(prediction[0][0]))
print(rounded_prediction )
```
